```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyOAuth

#creating the API authentication object 

scope = "user-library-read user-follow-read user-top-read playlist-read-private playlist-modify-private playlist-modify-public"

#insert the application auth settings, replacing the '#' characters
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id='######',
    client_secret='######',
    redirect_uri='######',
    scope=scope,
    requests_timeout=50
))
```

```python
#retrieving the features of the tracks I'm used to
#here, I considered my current top tracks and the saved songs

tracks_list_data = []

top_tracks = sp.current_user_top_tracks(limit=50)

for item in top_tracks['items']:
    track_dict = sp.audio_features(item['id'])[0]
    track_dict['song_name'] = item['name']
    tracks_list_data.append(track_dict)
    
saved_tracks = sp.current_user_saved_tracks(limit=50)

for item in saved_tracks['items']:
    track_dict = sp.audio_features(item['track']['id'])[0]
    track_dict['song_name'] = item['track']['name']
    tracks_list_data.append(track_dict)

df_tracks = pd.DataFrame(tracks_list_data)
```

```python
df_tracks
```

```python
df_tracks.drop_duplicates(inplace=True)
```

```python
df_tracks.describe()
```

```python
#get new recommendations for each track with Spotify recommender
recommendation_list = []

for index, row in df_tracks.iterrows():
    print(row['song_name'])
    recommendations = sp.recommendations(seed_tracks=[row['id']],limit=1)['tracks']
    
    while recommendations[0]['name'] in recommendations[0]['name'] in df_tracks['song_name']:
        print('Repeated recommendation. Generating a new one...')
        recommendations = sp.recommendations(seed_tracks=[row['id']],limit=1)['tracks']
    
    print('New recommendation!')
    
    for item in recommendations:
        recommendation_dict = sp.audio_features(item['id'])[0]
        print('Recommendation:',item['name'],'\n')
        recommendation_dict['song_name'] = item['name']
        recommendation_list.append(recommendation_dict)

df_recommended_tracks = pd.DataFrame(recommendation_list)
```

```python
#target the songs I already listen to as 1
#target new recommendations as 0

df_tracks['target']=1
df_recommended_tracks['target']=0
df_dataset = pd.concat([df_tracks, df_recommended_tracks])
df_dataset
```

```python
#Data preprocessing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df_dataset.drop(['type', 'id', 'uri', 'track_href', 'analysis_url', 'song_name', 'target'], axis=1)
y = df_dataset['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
#train a random forest classifier in order to try to classify the songs I already listen to and the new songs

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, StratifiedKFold

model = RandomForestClassifier(n_estimators=1000, max_depth=3, random_state=42)

model.fit(X_train, y_train)

preds = model.predict(X_test)

print(classification_report(y_test, preds))
```

##### These not so good metrics may be related to the fact that, not necessarily, a song targeted as 0 (not being part of my favorites, because it is a new recommendation) is not feature related to the songs I like, targeted as 1

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

model = LogisticRegression(C=0.01)

model.fit(X_train, y_train)

preds = model.predict(X_test)

print(classification_report(y_test, preds))
```

```python
#get a list of 3 recommendations for each song I'm used to listen to
#the idea is to use the trained model to predict the probability of this track being into the songs I commonly like

recommendation_list = []

for index, row in df_tracks.iterrows():
    print('---',row['song_name'],'---')
    aux = 0
    while aux!=3:
        recommendations = sp.recommendations(seed_tracks=[row['id']],limit=1)['tracks']

        while recommendations[0]['name'] in df_recommended_tracks['song_name'] or recommendations[0]['name'] in df_tracks['song_name']:
            print('Nova recomendação')
            recommendations = sp.recommendations(seed_tracks=[row['id']],limit=1)['tracks']

        print('Recomendou nova!')

        for item in recommendations:
            recommendation_dict = sp.audio_features(item['id'])[0]
            print('Recomendação:',item['name'],'\n')
            recommendation_dict['song_name'] = item['name']
            recommendation_list.append(recommendation_dict)
        
        aux+=1

df_new_tracks = pd.DataFrame(recommendation_list)
```

```python
#use the model to predict the probability of each song being part of the songs I like 
X_new = df_new_tracks.drop(['type', 'id', 'uri', 'track_href', 'analysis_url', 'song_name'], axis=1)
X_new = scaler.fit_transform(X_new)

df_new_tracks['probas'] = model.predict_proba(X_new)[:,1]
```

```python
#filter the tracks with probability > than 53% (arbitrary) of being one of my favorite songs
songs = df_new_tracks[df_new_tracks['probas'] > .53]['id'].values
songs
```
